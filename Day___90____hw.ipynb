```python
import os
import keras
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

train,test = keras.datasets.cifar10.load_data()
```

```python
image = train[0][0]
plt.imshow(image)
plt.show()
```

```python
gray = cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
plt.imshow(gray)
plt.show()
```

```python
hist = cv2.calcHist([gray], [0],None, [256],[0,256])
plt.figure()
plt.title("Gratscale Histogram")
plt.xlabel("Bins")
plt.ylabel("# of pixels")
plt.plot(hist)
plt.xlim([0,256])
plt.show()
```

```python
chans = cv2.split(image) #  3  channel 
colors = ("r", "g", "b")
plt.figure()
plt.title("'Flattened' Color Histogram")
plt.xlabel("Bins")
plt.ylabel("# of Pixels")

#  channel
for (chan, color) in zip(chans, colors):
    #  channel 
    hist = cv2.calcHist([chan], [0], None, [256], [0, 256])
 
    #  channel 
    plt.plot(hist, color = color)
    plt.xlim([0, 256])
plt.show()
```
