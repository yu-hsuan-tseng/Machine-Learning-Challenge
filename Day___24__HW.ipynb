```python
# import libraries

import os
import numpy as np
import pandas as pd
import copy,time
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
```

```python
f_train = os.path.expanduser('~/Desktop/titanic_train.csv')
f_test = os.path.expanduser('~/Desktop/titanic_test.csv')

df_train = pd.read_csv(f_train)
df_test = pd.read_csv(f_test)
```

```python
# setting up for feature engineering

train_y = df_train['Survived']
ids = df_test['PassengerId']
df_train = df_train.drop(['PassengerId','Survived'], axis=1)
df_test = df_test.drop(['PassengerId'], axis=1)
df = pd.concat([df_train,df_test])
df.head()
```

```python
# keep only object columns
obj_features = []
for dtype,feature in zip(df.dtypes,df.columns):
    if dtype == 'object':
        obj_features.append(feature)
print(f'{len(obj_features)}\n Numeric Features:{obj_features}\n')

df = df[obj_features]
df = df.fillna('None')
train_num = train_y.shape[0]
df.head()
```

```python
# survived data applies logistic regression better than linear regression
# label encoding + logistic regression

df_temp = pd.DataFrame()
for c in df.columns:
    df_temp[c] = LabelEncoder().fit_transform(df[c])
train_x = df_temp[:train_num]
estimator = LogisticRegression()
start = time.time()
print(f'shape: {train_x.shape}')
print(f'score: {cross_val_score(estimator, train_x, train_y, cv=5).mean()}')
print(f'time: {time.time() - start} sec')
```

```python
# one-hot encoding + logistic regression
df_temp = pd.get_dummies(df)
train_x = df_temp[:train_num]
estimator = LogisticRegression()
start = time.time()

print(f'shape :{train_x.shape}')
print(f'score:{cross_val_score(estimator,train_x,train_y,cv=5).mean()}')
print(f'time: {time.time() - start}sec')
```
